```python
import numpy as np
import matplotlib.pyplot as plt

# Set angular step in degrees
theta_step =1.13

# Set total number of steps
N_steps = 500000

x = []
y = []
theta = []

x.append(0.0)
y.append(0.0)
theta.append(0.0)

for i in range(N_steps):
    x_new = x[i] + np.cos((theta[i]+i*theta_step)*3.14159265/180.0)
    y_new = y[i] + np.sin((theta[i]+i*theta_step)*3.14159265/180.0)
    theta_new = theta[i]+i*theta_step
    x.append(x_new)
    y.append(y_new)
    theta.append(theta_new)

plt.figure(figsize=(12, 12))
plt.plot(x,y)

plt.show()



```
